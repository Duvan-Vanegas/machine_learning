# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
```

```
N=10
array_1 = np.arange(N) # Create an array of N values
array_2 = np.random.randint(low=-0,high=10,size=N) # Create an array of random int values
            # with the same lenght as array_1

print(array_1, array_2) # print both array_1 and array_2
print(array_1.shape, array_2.shape) # print both array_1 and array_2 shapes
print(np.sum(array_1), np.sum(array_2)) # print summation of array_1 and array_2
```

```
array_1[2] = 70 # Replace the value of array_1 in pos 3 by 70
print(array_1)
print(np.sum(array_1)) # Print summation again
```

```
result_array = array_1*array_2 # Apply element-wise multiplication among array_1 and array_2
print("Result element-wise multiplication: ", result_array) # Print results
```

### Going further

```
# Messing everything around.
img = cv2.imread('/content/drive/MyDrive/Machine Learning/machine_learning/Sem_01/lena.jpg')

# Invert the colors of the image
img_inverted = 255 - img

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 2)
plt.title('Inverted Image')
plt.imshow(cv2.cvtColor(img_inverted, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```
# Load the image in grayscale
img_gray = cv2.imread('/content/drive/MyDrive/Machine Learning/machine_learning/Sem_01/lena.jpg', cv2.IMREAD_GRAYSCALE)

# Modify the intensity of the pixels based on a condition
for r in range(img_gray.shape[0]):  # Go through rows
    for c in range(img_gray.shape[1]):  # Go through cols
        if img_gray[r, c] > 128:  # Example condition: if pixel intensity is greater than 128
            img_gray[r, c] = 255  # Set to white
        else:
            img_gray[r, c] = 0  # Set to black

# Display the modified image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

### Save an image using OpenCV

```
cv2.imwrite('newlena.png',img_gray)
```

### Crop the image - taking regions of interest (ROI)

```
# Extract some ROI
img = cv2.imread('/content/drive/MyDrive/Machine Learning/machine_learning/Sem_01/lena.jpg')

# Define the coordinates of the ROI (startY:endY, startX:endX)
# Example: Extracting a 100x100 pixel region starting at (50, 50)
roi = img[150:300, 150:300]

# Display the ROI
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```
img = cv2.imread('/content/drive/MyDrive/Machine Learning/machine_learning/Sem_01/lena.jpg')

# Convertir la imagen a escala de grises
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Definir las coordenadas del ROI (por ejemplo, un recuadro de 100x100 píxeles)
roi = img_gray[50:150, 50:150]

# Colocar el ROI en una nueva ubicación (por ejemplo, en la esquina superior izquierda)
img_gray[0:100, 0:100] = roi

# Mostrar la imagen resultante
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```
